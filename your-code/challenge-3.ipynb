## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
#sub 1

customer_totals = orders_df.groupby('CustomerID')['amount_spent'].sum().reset_index()
customer_totals.head(10)
```

```python
#sub 2

lower_quantile = 0.25
upper_quantile = 0.75

quantiles = customer_totals['amount_spent'].quantile([lower_quantile, upper_quantile])

selected_customers = customer_totals[
    (customer_totals['amount_spent'] >= quantiles[lower_quantile]) &
    (customer_totals['amount_spent'] <= quantiles[upper_quantile])
    ]

selected_customers.head(10)
```

```python
#sub 3

customer_totals['CustomerType'] = 'Preferred'
customer_totals.loc[
    (customer_totals['amount_spent'] >= quantiles[lower_quantile]) &
    (customer_totals['amount_spent'] <= quantiles[upper_quantile]), 'CustomerType'] = 'VIP'

customer_totals.head(10)
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here

jointed_df = pd.merge(orders_df, customer_totals, on='CustomerID', how='inner')
jointed_df.head(10)
```

```python
vip_customers_df = jointed_df[jointed_df['CustomerType'] == 'VIP']
vip_customer_counts = vip_customers_df.groupby('Country')['CustomerID'].nunique().reset_index()


most_vip_country = vip_customer_counts[vip_customer_counts['CustomerID'] == vip_customer_counts['CustomerID'].max()]
```

```python
most_vip_country.head(10)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
vip_and_preferred_df = vip_customers_df[(vip_customers_df['CustomerType'] == 'VIP') | (vip_customers_df['CustomerType'] == 'Preferred')]

customer_counts = vip_and_preferred_df.groupby('Country')['CustomerID'].nunique().reset_index()

most_combined_customers_country = customer_counts[customer_counts['CustomerID'] == customer_counts['CustomerID'].max()]

most_combined_customers_country.head(10)
```
